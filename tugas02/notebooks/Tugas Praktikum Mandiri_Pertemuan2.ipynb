```python
# Import library yang dibutuhkan
import pandas as pd
from sklearn.model_selection import train_test_split

# Membaca file day.csv yang sudah kamu upload tadi
df = pd.read_csv("day.csv")

# Menampilkan informasi dasar
print("Jumlah total data:", len(df))
print("\n5 data teratas:")
df.head()
```

```python
# Membagi dataset:
# - Data Training: 80%
# - Data Testing: 20%
train_full, test = train_test_split(df, test_size=0.20, random_state=42, shuffle=True)

# Data training penuh (train_full), ambil 10% sebagai data validation
train, validation = train_test_split(train_full, test_size=0.10, random_state=42, shuffle=True)

# Menampilkan jumlah data di setiap bagian
print("=== Jumlah Data ===")
print("Total data:", len(df))
print("Training:", len(train))
print("Validation:", len(validation))
print("Testing:", len(test))

# Memastikan jumlahnya sama dengan total data
print("Total semua subset:", len(train) + len(validation) + len(test))
```

```python
from IPython.display import display

print("=== DATA TRAINING (5 TERATAS) ===")
display(train.head())

print("=== DATA VALIDATION (5 TERATAS) ===")
display(validation.head())

print("=== DATA TESTING (5 TERATAS) ===")
display(test.head())
```

```python
import matplotlib.pyplot as plt

# Data proporsi
labels = ['Training', 'Validation', 'Testing']
values = [80, 10, 20]
colors = ['#A1C9F4', '#FFB482', '#8DE5A1']

plt.figure(figsize=(6,4))
bars = plt.bar(labels, values, color=colors, edgecolor='black')

# Tambahkan label nilai di atas batang
for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2, val + 1, f'{val}%',
             ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.title('Proporsi Dataset (Training, Validation, Testing)', fontsize=14, fontweight='bold')
plt.ylabel('Persentase (%)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
